```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib import pyplot

%matplotlib inline
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
hrr_gdf = pd.read_pickle('Dataframes/merged_hrr_gdf')
provider_gdf = pd.read_pickle('Dataframes/gdf_medicare_correct_hrr')
```

```python
hrr_gdf.head()
```

```python
provider_gdf.head()
```

```python
numerical_col = ['Total Discharges', 'Average Covered Charges', 'Average Total Payments', 'Average Medicare Payments']
numerical_dict = dict.fromkeys(numerical_col, 'mean')
```

```python
provider_agg = provider_gdf.groupby(['Provider Name']\
                                            ,as_index=False).agg(numerical_dict)

provider_geo_agg = provider_gdf[['Provider Name','geometry', 'HRRD' ]].\
                    drop_duplicates(subset = ['Provider Name'])
```

```python
provider_gdf_agg= provider_geo_agg.merge(provider_agg,\
                           on = 'Provider Name')
```

```python
provider_gdf_agg.head()
```

```python
plt.rc('legend',fontsize=14)
```

```python
fig, (ax1) = pyplot.subplots(ncols=1)
fig.set_size_inches(25, 12, forward=True)
#plt.rcParams["figure.figsize"] = (20,15)
hrr_gdf.plot(ax=ax1, column='Average Covered Charges', scheme='fisher_jenks',\
                   cmap= plt.cm.Reds, legend=True, k =6, label = 'lol')
ax1.get_legend().set_bbox_to_anchor((1,0.5))
ax1.get_legend().set_title('Cost In Dollars', prop={'size':14})
fig.suptitle('Average Covered Charges by HRR', fontsize=20)
#g._legend.set_title(new_title)
#ax1.set_xlim([-130, -60])
#ax1.set_ylim([23, 52])
#ax1.tick_params(labelsize=10)
ax1.set_axis_off()
plt.savefig('Vis/test')
```

```python
ax2 = provider_gdf_agg.plot(column='Average Covered Charges', scheme='fisher_jenks', cmap= plt.cm.Reds, legend=True, k =6)
ax2.set_xlim([-130, -60])
ax2.set_ylim([23, 52])
plt.rcParams["figure.figsize"] = [20,15]
ax2.tick_params(labelsize=10) 
ax2.legend(loc = 4)
```

```python
plt.rcParams['axes.titlepad'] = 20 
base = hrr_gdf.plot(column='Average Covered Charges', scheme='fisher_jenks', \
                    cmap= plt.cm.Blues, legend=True, k =6, \
                   label = 'HRR Average Covered Charges')
base.grid(False)

base.get_legend().set_title('ACC by HRR in Dollars Amount', prop={'size':12})
leg = base.get_legend()
ax3 = provider_gdf_agg.plot(ax = base, column='Average Covered Charges', \
                            scheme='fisher_jenks', cmap= plt.cm.Reds, \
                            markersize = 8, legend=True, k =8)
ax3.grid(False)
ax3.add_artist(leg)
leg.set_bbox_to_anchor((1, 1))
#ax3.get_legend().set_bbox_to_anchor((1,0.5))
ax3.get_legend().set_title('ACC by Indivdual Provider in Dollars Amount', prop={'size': 12})
ax3.set_xlim([-130, -60])
ax3.set_ylim([23, 52])
plt.title('Average Covered Charges(ACC) by HRR and Individual Providers', fontsize=20)
plt.rcParams["figure.figsize"] = [25,18]
```
